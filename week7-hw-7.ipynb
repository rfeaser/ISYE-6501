## Question 15.2

In the videos, we saw the “diet problem”. (The diet problem is one of the first large-scale optimization problems to be studied in practice.
```python
#Creating the problem variable to contain the problem data
prob = LpProblem("Min_Diet_Cost", LpMinimize)
```

```python
foods = diet['Foods'].tolist()

#Dictionary of cost of each food
price = dict(zip(diet['Foods'], diet['Price/ Serving']))

#Dictionary of serving size of each food
serving = dict(zip(diet['Foods'], diet['Serving Size']))

#Dictionary of calories for each serving size of food
calories = dict(zip(diet['Foods'], diet['Calories']))

#Dictionary of cholesterol for each serving size of food
cholesterol = dict(zip(diet['Foods'], diet['Cholesterol mg']))

#Dictionary of total fat for each serving size of food
fat = dict(zip(diet['Foods'], diet['Total_Fat g']))

#Dictionary of sodium for each serving size of food
sodium = dict(zip(diet['Foods'], diet['Sodium mg']))

#Dictionary of carbohydrates for each serving size of food
carbs = dict(zip(diet['Foods'], diet['Carbohydrates g']))

#Dictionary of dietary fiber for each serving size of food
fiber = dict(zip(diet['Foods'], diet['Dietary_Fiber g']))

#Dictionary of protein for each serving size of food
protein = dict(zip(diet['Foods'], diet['Protein g']))

#Dictionary of vitamin A for each serving size of food
vit_a = dict(zip(diet['Foods'], diet['Vit_A IU']))

#Dictionary of vitamin C for each serving size of food
vit_c = dict(zip(diet['Foods'], diet['Vit_C IU']))

#Dictionary of calcium for each serving size of food
calcium = dict(zip(diet['Foods'], diet['Calcium mg']))

#Dictionary of iron for each serving size of food
iron= dict(zip(diet['Foods'], diet['Iron mg']))
```

```python
# Creating a dictionary called 'food_vars' to contain the referenced Variables
food_vars = LpVariable.dicts("Foods", foods, 0)
```

```python
# The objective function is added to 'prob' first
prob += lpSum([price[i] * food_vars[i] for i in foods])
```

```python
# The min/max constraints are added to 'prob'

#Calories
prob += lpSum([calories[i] * food_vars[i] for i in foods]) >= 1500, 'Calorie Min'
prob += lpSum([calories[i] * food_vars[i] for i in foods]) <= 2500, 'Calorie Max'

#Cholesterol
prob += lpSum([cholesterol[i] * food_vars[i] for i in foods]) >= 30, 'Cholesterol Min'
prob += lpSum([cholesterol[i] * food_vars[i] for i in foods]) <= 240, 'Cholesterol Max'

#Total fat
prob += lpSum([fat[i] * food_vars[i] for i in foods]) >= 20
prob += lpSum([fat[i] * food_vars[i] for i in foods]) <= 70

#Sodium
prob += lpSum([sodium[i] * food_vars[i] for i in foods]) >= 800
prob += lpSum([sodium[i] * food_vars[i] for i in foods]) <= 2000

#Carbs
prob += lpSum([carbs[i] * food_vars[i] for i in foods]) >= 130
prob += lpSum([carbs[i] * food_vars[i] for i in foods]) <= 450

#Fiber
prob += lpSum([fiber[i] * food_vars[i] for i in foods]) >= 125
prob += lpSum([fiber[i] * food_vars[i] for i in foods]) <= 250

#Protein
prob += lpSum([protein[i] * food_vars[i] for i in foods]) >= 60
prob += lpSum([protein[i] * food_vars[i] for i in foods]) <= 100

#Vitamin A
prob += lpSum([vit_a[i] * food_vars[i] for i in foods]) >= 1000
prob += lpSum([vit_a[i] * food_vars[i] for i in foods]) <= 10000

#Vitamin C
prob += lpSum([vit_c[i] * food_vars[i] for i in foods]) >= 400
prob += lpSum([vit_c[i] * food_vars[i] for i in foods]) <= 5000

#Calcium
prob += lpSum([calcium[i] * food_vars[i] for i in foods]) >= 700
prob += lpSum([calcium[i] * food_vars[i] for i in foods]) <= 1500

#Iron
prob += lpSum([iron[i] * food_vars[i] for i in foods]) >= 10
prob += lpSum([iron[i] * food_vars[i] for i in foods]) <= 40
```

```python
# solve the optimization problem!
prob.solve()
```

```python
# print the foods of the optimal diet
print('Optimization Solution:')
for var in prob.variables():
    if var.varValue > 0:
        if str(var).find('Chosen'):
            print(str(var.varValue) + " units of " + str(var))
            
# print the costs of the optimal diet             
print("Total cost of food = $%.2f" % value(prob.objective))
```

Two columns have the same name energy. Let's fix this so that way we're able to apply constraints to both.

```python
class renamer():
    def __init__(self):
        self.d = dict()
    def __call__(self, x):
        if x not in self.d:
            self.d[x] = 0
            return x
        else:
            self.d[x] += 1
            return "%s_%d" % (x, self.d[x])
                
diet2 = diet2.rename(columns=renamer())
```
